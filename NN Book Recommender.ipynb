```python
import numpy as np
import pandas as pd
```

```python
books=pd.read_csv('BX-Books.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
books=books[['ISBN','Book-Title','Book-Author']]
```

```python
books.columns=['ISBN','Book Name','Author']
```

```python
books.shape
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.columns=['user_id','ISBN','rating']
```

```python
type(ratings['ISBN'])
```

```python
user_min=ratings['user_id'].value_counts()
user_min=user_min[user_min>200]
user_min
```

```python
ratings=ratings[ratings['user_id'].isin(user_min.index)]
len(ratings)
```

```python
ratings.head(1)
```

```python
ratings_with_bookname=ratings.merge(books,on='ISBN')
```

```python
final_total_rating=ratings_with_bookname.groupby('Book Name')['rating'].agg('count').reset_index().rename(columns={'rating':'num_rating'})
```

```python
final_total_rating.head(70)
```

```python
ultimate_final=ratings_with_bookname.merge(final_total_rating,on='Book Name')
```

```python
ultimate_final.head()
```

```python
ultimate_final=ultimate_final[ultimate_final['num_rating']>=50]
len(ultimate_final)
```

```python
ultimate_final=ultimate_final.drop_duplicates(['user_id','Book Name'])
```

```python
ultimate_final.head(20)
```

```python
ultimate_pivot=ultimate_final.pivot_table(values='rating',index='Book Name',columns='user_id').fillna(0)
```

```python
ultimate_pivot
```

```python
from scipy.sparse import csr_matrix
ultimate_sparse=csr_matrix(ultimate_pivot)
```

```python
ultimate_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
nn=NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
nn.fit(ultimate_sparse)
```

```python
distance,suggestion=nn.kneighbors(ultimate_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestion=suggestion.ravel()
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    if i==0:
        print("Searched for:",ultimate_pivot.index[suggestion[i]])
    else:
        print(ultimate_pivot.index[suggestion[i]])
```

```python
np.where(ultimate_pivot.index=='Animal Farm')[0][0]
```

```python
def suggest(book_name):
    bindex=np.where(ultimate_pivot.index==book_name)[0][0]
    distance,suggestion=nn.kneighbors(ultimate_pivot.iloc[bindex,:].values.reshape(1,-1),n_neighbors=6)
    suggestion=suggestion.ravel()
    for i in range(len(suggestion)):
        if i==0:
            print("Searched for:",ultimate_pivot.index[suggestion[i]])
        else:
            print(ultimate_pivot.index[suggestion[i]])
```

```python
suggest('Animal Farm')
```
