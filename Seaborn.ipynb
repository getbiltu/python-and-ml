```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='hex')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
sns.barplot(x='smoker',y='tip',data=tips)
```

```python
sns.barplot(x='smoker',y='tip',data=tips,estimator=np.std)
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```
