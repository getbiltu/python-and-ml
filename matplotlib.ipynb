```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 2D PLOT

```python
product_id=[1,2,3,4,5,6,7]
price=[345,120,400,333,124,900,100]
quantity=[106,257,145,123,156,160,130]
discount=[13,24,40,50,25,30,10]
```

```python
plt.plot(product_id,price,color='orange',linewidth=5,linestyle='dotted')
plt.xlabel('Product')
plt.ylabel('Price')
plt.title('Price Comparision')
```

```python
plt.plot(product_id,price,color='orange',linewidth=2,marker='D')
plt.xlabel('Product')
plt.ylabel('Price')
plt.title('Price Comparision')
```

```python
#Many Plot In Single Graph
plt.plot(product_id,price,color='orange',linewidth=2,marker='D',label='Price')
plt.plot(product_id,quantity,color='green',marker='+',label='Quantity')
plt.plot(product_id,discount,color='black',marker='.',label='Discount')
plt.xlabel('Product')
plt.legend()
plt.title('Price Description')
```

# Bargraph

```python
year=[2012,2013,2014,2015,2016,2017,2018]
offline_price=[1200,1351,1100,1004,1300,1500,1541]
online_price=[110,1200,1400,1299,1111,1390,789]
```

```python
plt.bar(year,offline_price,color='orange',width=.5)
plt.bar(year,online_price,color='violet',width=.5)
plt.title('Year by Year Prices')
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017,2018]
year=np.array(year)
offline_price=[1200,1351,1100,1004,1300,1500,1541]
online_price=[110,1200,1400,1299,1111,1390,789]
plt.bar(year+.2,offline_price,color='orange',width=.4)
plt.bar(year-.2,online_price,color='red',width=.4)
plt.title('Year by Year Prices')
plt.show()
```

# Histograms

```python
price=[30,35,105,55,65,10,46,78,87,90,36,76,120,97,56,88,21,33,105]
price2=[40,56,12,34,56,68,78,102,120,34,56,68,89,90,12,46,76,23,67,89]
```

```python
plt.hist(price,bins=[30,60,90,120],rwidth=.9,color='orange')
plt.show()
```

```python
plt.hist([price,price2],bins=[30,60,90,120],rwidth=.5,color=['orange','red'])
plt.show()
```

# Pie-charts

```python
area=['Marketing','Salery','Ads','Office Rent','Sales']
budget=[23,34,12,67,40]
```

```python
plt.pie(budget,labels=area,autopct='%.2f%%',explode=[0,0,.2,0,0])
plt.show()
```
