```python
import os
from collections import Counter
```

```python
folder='email/'
```

```python
files=os.listdir(folder)
```

```python
len(files)
```

```python
emails=[folder + file for file in files]
emails
```

```python
words=[]
for email in emails:
    f=open(email,encoding='latin-1')
    blob=f.read()
    words+=blob.split(" ")
```

```python
len(words)
```

```python
words[0:5]
```

```python
for i in range(len(words)):
    if not words[i].isalpha():
        words[i]=""
```

```python
word_dictionary=Counter(words)
del word_dictionary[""]
word_dictionary=word_dictionary.most_common(10000)
```

```python
word_dictionary[9:11]
```

```python
len(word_dictionary)
```

```python
feature_set=[]
labels=[]
for email in emails:
    f=open(email,encoding='latin-1')
    blob=f.read().split(" ")
    data=[]
    for i in word_dictionary:
        data.append(blob.count(i[0]))
    
    feature_set.append(data)
    if 'ham' in email:
        labels.append(0)
    if 'spam' in email:
        labels.append(1)
```

```python
len(feature_set)
```

```python
import numpy as np
feature_set=np.array(feature_set)
labels=np.array(labels)
```

```python
feature_set.shape
labels.shape
```

```python
feature_set
```

```python
from sklearn.cross_validation import train_test_split

X_train,X_test,y_train,y_test=train_test_split(feature_set,labels,test_size=0.2,random_state=20)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_test[562]
clf.predict(X_test[562].reshape(1,-1))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
